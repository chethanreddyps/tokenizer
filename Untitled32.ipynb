```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

model = SentenceTransformer('all-MiniLM-L6-v2')

text_1 = "chethanis my name"
text_2 = "my name is chethan"

embeddings = model.encode([text_1, text_2])

similarity_score = cosine_similarity(
    [embeddings[0]],
    [embeddings[1]]
)[0][0]

print("Similarity Score:", similarity_score)
```

```python
from transformers import pipeline

sentiment_analyzer = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)

texts = [
    "I love this product, it works perfectly!",
    "This is the worst experience I have ever had.",
    "The movie was okay, not great but not bad."
]

results = sentiment_analyzer(texts)

for text, result in zip(texts, results):
    print(f"Text: {text}")
    print(f"Sentiment: {result['label']}, Confidence: {result['score']:.3f}")
    print("-" * 50)
```

```python
# pip install kagglehub[pandas-datasets] transformers torch scikit-learn

import kagglehub
from kagglehub import KaggleDatasetAdapter
import torch
from transformers import AutoTokenizer, AutoModel
from sklearn.metrics.pairwise import cosine_similarity

# Load dataset
df = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "saurabhshahane/fake-news-classification",
    "WELFake_Dataset.csv"
).dropna()

# Load Hugging Face model
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")
model.eval()

# Mean pooling embedding
def embed(texts):
    tokens = tokenizer(texts, padding=True, truncation=True, max_length=256, return_tensors="pt")
    with torch.no_grad():
        output = model(**tokens).last_hidden_state
    mask = tokens["attention_mask"].unsqueeze(-1)
    return (output * mask).sum(1) / mask.sum(1)

# Take two news articles
t1, t2 = df.iloc[0]["text"], df.iloc[1]["text"]

# Similarity
emb = embed([t1, t2])
score = cosine_similarity(emb[0].unsqueeze(0), emb[1].unsqueeze(0))[0][0]

print("Similarity score:", round(score, 3))
```

```python
# Fake news example
fake_news = """
NASA confirms Earth will face 15 days of complete darkness next month
due to a rare planetary alignment, scientists warn.
"""

# Take a real news article from dataset
real_news = df[df["label"] == 1].iloc[0]["text"]  # label 1 = REAL

# Embed texts
embeddings = embed([fake_news, real_news])

# Compute similarity
similarity_score = cosine_similarity(
    embeddings[0].unsqueeze(0),
    embeddings[1].unsqueeze(0)
)[0][0]

print("Fake News Text:\n", fake_news.strip(), "\n")
print("Similarity with REAL news:", round(similarity_score, 3))

# Simple interpretation
if similarity_score > 0.7:
    print("⚠️ High semantic similarity (possible paraphrase or misleading framing)")
else:
    print("✅ Low similarity (likely unrelated or fabricated content)")
```
